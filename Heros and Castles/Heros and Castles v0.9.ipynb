```python
x_button = [1346,299]
buy_button = [1171,788]
battle_academy = [736,430]
buildings_button = [709,971]
units_button = [1102,974]
peons_left = [177,321]
wall_left = [176,444]
any_click = [1327,737]
play_button = [1351,939]
one_button = [589,367]
next_button = [1345,940]
accept_go = [1059,657]

start_coins = 50
gold_mine_cost = 15
battel_academy_cost = 125
peons_cost = 40

game_time = 180

def buffer(t=3):
    time.sleep(t)

time.sleep(10)
    
while True:
    pag.click(play_button)
    buffer(2)
    pag.click(accept_go)
    buffer(6)
    pag.click(any_click)
    buffer(0.15)
    start_time = time.time()
    buffer(2)

    coins = start_coins+2
    # buy first gold mine
    pag.click(buildings_button)
    buffer(1)
    pag.click(buy_button)
    buffer(1)
    pag.click(x_button)
    buffer(1)
    pag.click(x_button)
    buffer(1)
    coins-=(gold_mine_cost-2)

    # wait until 40 for first peons
    while True:
        if coins>43:
            break  
        time.sleep(1)
        coins+=2

    # buy first peons
    pag.click(units_button)
    buffer(1)
    pag.click(buy_button)
    buffer(1)
    pag.click(x_button)
    buffer(1)
    pag.click(x_button)
    buffer(1)
    coins-=(peons_cost-2)


    # wait until 125 to buy academy
    while True:
        if coins>128:
            break  
        time.sleep(1)
        coins+=2


    pag.click(buildings_button)
    buffer(1)
    pag.click(battle_academy)
    buffer(1)
    pag.click(buy_button)
    buffer(1)
    pag.click(x_button)
    buffer(1)
    pag.click(x_button)
    buffer(1)
    coins-=(battel_academy_cost-2)


    # wait to buy more peons
    while True:
        if coins>43:
            pag.click(units_button)
            buffer(1)
            pag.click(buy_button)
            buffer(1)
            pag.click(x_button)
            buffer(1)
            pag.click(x_button)
            buffer(1)
            coins-=(peons_cost-2)
            break
        time.sleep(1)
        coins+=2

    while True:
        if coins>43:
            pag.click(units_button)
            buffer(1)
            pag.click(buy_button)
            buffer(1)
            pag.click(x_button)
            buffer(1)
            pag.click(x_button)
            buffer(1)
            coins-=(peons_cost-2)
            break
        time.sleep(1)
        coins+=2

    while True:
        if time.time()-start_time>game_time:
            break

        if coins>43 and time.time()-start_time<game_time:
            pag.click(units_button)
            buffer(1)
            pag.click(buy_button)
            buffer(1)
            pag.click(x_button)
            buffer(1)
            pag.click(x_button)
            buffer(1)
            coins-=(peons_cost-2)
            break
        time.sleep(1)
        coins+=2
    while True:
        if time.time()-start_time>game_time:
            buffer(30)
            pag.click(any_click)
            buffer(3)
            pag.click(one_button)
            buffer(1)
            pag.click(next_button)
            buffer(3)
            break

```
